## Big-O Notation.

To characterize the efficiency of an algorithm in terms of execution time we need to quantify the number of steps the algorithm will require -- with each step being the basic unit of computation.

**Order of magnitude function (Big-O)** describes the $T(n)$ time it takes to solve a problem of size $n$ that grows the fastest as the value of $n$ increases, written as $O(f(n))$ and is an approximation for the actual number of steps in the computation. Where $f(n)$ is the dominant part of $T(n)$.

| $f(n)$ | Name | 
| --- | --- |
| $1$ | Constant |
| $\log(n)$ | Logarithmic |
| $n$ | Linear |
| $n\log(n)$ | Log Linear |
| $n^{2}$ | Quadratic |
| $n^{3}$ | Cubic |
| $2^{n}$ | Exponential |
